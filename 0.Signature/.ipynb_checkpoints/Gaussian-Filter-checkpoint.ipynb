```python
import numpy as np
import pandas as pd
from scipy import ndimage
import random
import os
import matplotlib.pyplot as plt
import scanpy as sc
```

```python
sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_header()
sc.settings.set_figure_params(dpi=200, facecolor='white')
```

```python
data = pd.read_csv('../2.Nebula/result/t_CD8-CXCL13_nebula_result_mb.csv')
```

```python
data
```

```python
data_filter = ndimage.gaussian_filter(data['log2FC'],sigma=10)
```

```python
data_filter
```

```python
data_filter = pd.DataFrame(data_filter)
data_filter.rename(columns={0:'number'},inplace=True)
```

```python
data_filter
```

```python
fig = plt.figure(figsize = (10,8))
```

```python
plt.plot(data.index, # x轴数据
         data.coef, # y轴数据
         #linestyle = '-', # 折线类型
         #linewidth = 0.5, # 折线宽度
         color = 'red', # 折线颜色
         marker = 'o', # 点的形状
         markersize = 0, # 点的大小
         markeredgecolor='black', # 点的边框色
         markerfacecolor='brown') # 点的填充色

plt.plot(data_filter.index, # x轴数据
         data_filter.number, # y轴数据
         #linestyle = '-', # 折线类型
         #linewidth = 0.5, # 折线宽度
         color = 'blue', # 折线颜色
         marker = 'o', # 点的形状
         markersize = 0, # 点的大小
         markeredgecolor='black', # 点的边框色
         markerfacecolor='brown') # 点的填充色
```

```python
diff2 = pd.DataFrame(np.diff(data_filter.number,n=2))
```

```python
diff2.rename(columns={0:'diff2'},inplace=True)
```

```python
plt.plot(diff2.index, # x轴数据
         diff2.diff2, # y轴数据
         linestyle = '-', # 折线类型
         linewidth = 0.1, # 折线宽度
         color = 'red', # 折线颜色
         marker = 'o', # 点的形状
         markersize = 0, # 点的大小
         markeredgecolor='black', # 点的边框色
         markerfacecolor='brown') # 点的填充色
```

```python
final = ndimage.gaussian_filter(diff2.diff2, sigma=10)
final = pd.DataFrame(final)
final.rename(columns={0:'number'}, inplace=True)
```

```python
final.head(173)
```

```python

plt.plot(diff2.head(200).index, # x轴数据
         diff2.head(200).diff2, # y轴数据
         linestyle = '-', # 折线类型
         linewidth = 0.5, # 折线宽度
         color = 'red', # 折线颜色
         marker = 'o', # 点的形状
         markersize = 0, # 点的大小
         markeredgecolor='black', # 点的边框色
         markerfacecolor='brown') # 点的填充色

plt.plot(final.head(200).index, # x轴数据
         final.head(200).number, # y轴数据
         linestyle = '-', # 折线类型
         linewidth = 0.5, # 折线宽度
         color = 'blue', # 折线颜色
         marker = 'o', # 点的形状
         markersize = 0, # 点的大小
         markeredgecolor='black', # 点的边框色
         markerfacecolor='brown') # 点的填充色

```
