```python
import numpy as np
import pandas as pd
import scanpy as sc
```

```python
sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_header()
sc.settings.set_figure_params(dpi=200, facecolor='white')
```

```python
subcluster = 't_CD8-CXCL13'
```

```python
adata = sc.read_h5ad('./data/Tcell_PDL1.h5ad')
```

```python
adata = adata[adata.obs['Sub_Cluster']==subcluster]
```

```python
adata
```

```python
ann = pd.read_csv('./result/cluster/' + subcluster + '_cluster.csv',
                  index_col=0,header=0,sep=',',
                 usecols=['cell','cluster'])
```

```python
adata.obs = adata.obs.join(ann,how='left')
```

```python
adata.obs
```

```python
df = adata.obs[['Patient', 'Response','Timeline']].value_counts().reset_index(name='count')[['Patient','Response','Timeline']]
df
```

```python
for i in df.index:
    df.loc[i, 'count'] = len(adata[(adata.obs['Patient']==df.loc[i, 'Patient'])&(adata.obs['Timeline']==df.loc[i, 'Timeline'])])
df
```

```python
adata.obs['cluster'] = adata.obs['cluster'].replace((12,13,16),0)
```

```python
df['Res_Time'] = df.apply(lambda x: x['Response']+ '_'+ x['Timeline'],axis=1)
```

```python
for i in df.index:
    df.loc[i, 'fraction'] = len(adata[(adata.obs['Patient']==df.loc[i,'Patient'])&(adata.obs['Timeline']==df.loc[i,'Timeline'])&(adata.obs['cluster']==0)])/df.loc[i,'count']
df
```

```python
for i in df.index:
    df.loc[i,'fraction'] = df.loc[i, 'fraction'] * 100
df
```

```python
df.to_csv('./data/lineplot/'+subcluster+'lineplot.csv',header=True,index=True)
```

```python
df2 = df[['Response','Timeline']].value_counts().reset_index(name='count')[['Response','Timeline']]
df2
```

```python
for i in df2.index:
    df2.loc[i, 'mean'] = np.mean(df[(df['Response']==df2.loc[i, 'Response'])&(df['Timeline']==df2.loc[i, 'Timeline'])]['fraction'])
```
