```python
import numpy as np
import pandas as pd
import scanpy as sc
import anndata as ad
import os
```

```python
sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_header()
sc.settings.set_figure_params(dpi=200, facecolor='white')
```

```python
#读取含有注释信息的细胞数据
dt_dir = '/Users/liupeiwen/BC/New_analysis/1.Toomanycells_result/Tcell/data/'
adata = sc.read_h5ad(dt_dir+'Tcell_PDL1.h5ad')
```

```python
subcluster = 't_CD8-CXCL13'
```

```python
# 选取细胞类型
adata = adata[adata.obs['Sub_Cluster']==subcluster]
adata
```

```python
adata.to_df()
```

```python
ann = pd.read_csv('/Users/liupeiwen/BC/New_analysis/1.Toomanycells_result/Tcell/result/cluster/t_CD8-CXCL13_cluster.csv',
                  index_col=0,header=0,sep=',',
                 usecols=['cell','cluster'])
```

```python
adata.obs = adata.obs.join(ann,how='left')
```

```python
adata.obs
```

```python
adata.obs['cluster'] = adata.obs['cluster'].replace((12,13,16),0)
```

```python
# 设置case组和control组
for i in adata.obs_names:
    if adata.obs.loc[i,'cluster']==0:
        adata.obs.loc[i,'divide']='NR_Enrich'
    else:
        adata.obs.loc[i,'divide']='R_Enrich'
```
